# Exploratory Data Analysis (EDA)

```python
# Count of passengers by class
print(df['Class'].value_counts())
```

```python
# Count of passengers by survival status
print(df['Survived'].value_counts())
```

```python
# Visualizing the survival count
sns.countplot(x='Survived', data=df, palette='Blues')
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
# Visualizing age distribution
sns.histplot(df['Age'].dropna(), bins=20, kde=True, color='green')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Filter the dataset to include only valid passenger classes (1, 2, 3)
df_passengers = df[df['Class'].isin([1, 2, 3])]

# Count the occurrences of each class (1, 2, 3)
class_counts_passengers = df_passengers['Class'].value_counts()

# Calculate the percentage for each class
class_percentages_passengers = (class_counts_passengers / len(df_passengers)) * 100

# Print the class percentages for passengers only
print(class_percentages_passengers)
```
